```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset ('titanic')
df.head()
```

# Count Plot

```python
sns.countplot(data = df,
              x="class"
              )
```

```python
sns.countplot(data = df,
              x="class",
              hue="survived")
```

```python
sns.countplot(data = df,
              x="class",
              hue="survived",
              stat="percent"
              )
```

# Line PLot


```python
df = sns.load_dataset("flights")
df.head()
```

```python
may_flights = df.query("  month == 'May'   ")

sns.lineplot(data=may_flights,
             x="year",
             y="passengers")
```

```python
df_wide = df.pivot(index="year", columns="month", values="passengers")
df_wide.head()
```

```python
sns.lineplot(data=df_wide["May"])
```

```python
sns.lineplot(data=df_wide)
```

```python
sns.lineplot(data=df,
             x="year",
             y="passengers"
             )
```

```python
sns.lineplot(data=df,
             x="year",
             y="passengers",
             hue="month")
```

# Pie Plot

```python
# declaring data
data_values = [44, 45, 40, 41, 39]
keys = ['Class 1', 'Class 2', 'CLass 3', 'Class 4', 'Class 5']

# declaring exploding pie
exp_list = [0, 0.1, 0, 0, 0]
# define Seaborn color palette to use


# plotting data on chart
plt.pie(  x = data_values,
          labels=keys,
          explode=exp_list,
          autopct='%.0f%%'
          )

# displaying chart
plt.show()
```
